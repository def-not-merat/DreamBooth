```python
# This code is GPT generated. i used to to visualise my images in to pretty looking grids. for my report

from PIL import Image, ImageDraw, ImageFont, ImageOps
import os
import math

# Path to the root directory containing image subfolders
root_folder = '../experiments/misha_no_prior/'  # <-- Change this to your actual path
# root_folder = 'db/misha_db/'
pic_name = "no_prior"
label = "no prior loss"

# Font settings
font = ImageFont.load_default()
desc_height = 30  # Space for row description
target_size = (256, 256)  # Resize all images to this size

# Step 1: Collect images grouped by subfolder
image_rows = []     # List of image rows
descriptions = []   # Subfolder names = row labels

for dirpath, _, filenames in sorted(os.walk(root_folder)):
    image_paths = sorted([
        os.path.join(dirpath, f)
        for f in filenames
        if f.lower().endswith(('.png', '.jpg', '.jpeg'))
    ])
    if len(image_paths) == 5:
        images = [ImageOps.exif_transpose(Image.open(p)).resize(target_size) for p in image_paths]
        image_rows.append(images)
        descriptions.append(os.path.basename(os.path.dirname(dirpath)))

# Step 2: Image and grid sizes
image_width, image_height = target_size
cols = 5
rows = len(image_rows)
grid_width = cols * image_width
grid_height = rows * (image_height + desc_height)

# Step 3: Create the final grid image
grid = Image.new('RGB', (grid_width, grid_height), color=(255, 255, 255))
draw = ImageDraw.Draw(grid)

for row_idx, (images, desc) in enumerate(zip(image_rows, descriptions)):
    # Draw description
    y_desc = row_idx * (image_height + desc_height)
    bbox = draw.textbbox((0, 0), label, font=font)
    text_width = bbox[2] - bbox[0]
    text_x = (grid_width - text_width) // 2
    draw.text((text_x, y_desc), label, font=font, fill=(0, 0, 0))

    # Paste images
    for col_idx, img in enumerate(images):
        x = col_idx * image_width
        y = y_desc + desc_height
        grid.paste(img, (x, y))

# Save and/or show
# grid.save(f'image_grid_by_folder_resized_{target_size[0]}.png')
# print(f"../results/{pic_name}.png")
grid.save(f"../results/{pic_name}.png")
# grid.show()

```
